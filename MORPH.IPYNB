```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread('Mapache.jpeg', cv.IMREAD_GRAYSCALE)
_, binary_img = cv.threshold(img, 127, 255, cv.THRESH_BINARY)
plt.imshow(binary_img, cmap='gray')
plt.show()
```

```python
UMBRAL = 128
img_bn = cv.threshold(img, UMBRAL, 255, cv.THRESH_BINARY)[1]
```

```python
np.unique(img)
```

## Erode

```python
s_e = np.ones((3,3), dtype=np.uint8)
s_e2 = np.ones((5,5), dtype=np.uint8)
```

```python
cv.erode(img_bn, s_e)
```

```python
plt.imshow(np.hstack((img_bn, cv.erode(img_bn, s_e, iterations=2))),"gray")
plt.show()
```

```python
plt.imshow(np.hstack((img_bn, cv.erode(img_bn, s_e2, iterations=1))),"gray")
plt.show()
```

## Dilatacin

```python
plt.imshow(cv.dilate(img_bn, s_e, iterations=2), "gray")
```

```python
plt.imshow(cv.dilate(img_bn, s_e2, iterations=1), "gray")
```
