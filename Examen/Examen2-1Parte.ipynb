## Transformación negativa

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img = cv.imread("Mapache.jpeg",0)
plt.imshow(img, cmap="gray")
```

```python
plt.imshow(np.bitwise_not(img,),"gray")
```

## Transformación logarítmica

```python
m = np.max(img)
c = 255/(np.log(1+m))
alto, ancho = img.shape
s = np.zeros(img.shape)
for x in range(alto):
    for y in range (ancho):
        r = img[x,y]
        s[x,y]=c*np.log(1)+r
plt.imshow(s, cmap= "gray")
```

## Transformación gamma, aclarar la imagen

```python
m = np.max(img)
pot =  0.05

alt, anch = img.shape
s =np.zeros(img.shape)
for x in range(alt):
    for y in range(anch):
        r = img[x,y]
        s[x,y] = (1)+r**pot

plt.imshow(s, cmap="gray")
```

## Transformación gamma, oscurecer la imagen

```python
m = np.max(img)
pot =  25

alt, anch = img.shape
s =np.zeros(img.shape)
for x in range(alt):
    for y in range(anch):
        r = img[x,y]
        s[x,y] = (1)+r**pot

plt.imshow(s, cmap="gray")
```
