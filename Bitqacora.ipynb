```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv.imread("images\coins.png", cv.IMREAD_GRAYSCALE)
plt.imshow(img, "gray")
```

```python
np.unique(img)
```

```python
UMBRAL = 128 #umbral es variable, segun cv puede ser float pero si img es uint8 mandar uint
cv.threshold(img, UMBRAL, 255, cv.THRESH_BINARY)[1] #max val para ver si va a ser 0, 1 o 0, 255
img_bin = cv.threshold(img, UMBRAL, 255, cv.THRESH_BINARY)[1]
plt.imshow(img_bin, "gray")
```

```python
s_e = np.ones([3,3], dtype= np.uint8) #"kernel"
s_e2= np.ones([5,5], dtype= np.float32)
```

```python
cv.erode(img_bin, s_e)
img_eroded = cv.erode(img_bin, s_e)
plt.imshow(img_eroded, "gray") #primer plano es blanco, quita info de blanco y lo hace negro
```

```python
plt.subplot(121); plt.imshow(img_bin, cmap='gray'); plt.title("Original") 

plt.subplot(122); plt.imshow(img_eroded, cmap='gray'); plt.title("Erosion")
```

```python
img_erodedx3 = cv.erode(img_bin, s_e, iterations=3) #erosiona, a la resultante erosiona de nuevo y a la resultante la erosiona de nuevo

plt.imshow(np.hstack((img_bin, img_eroded, img_erodedx3)), "gray")
```

```python
img_SE5= cv.erode(img_bin, s_e2)
img_SE5_erodedx3 = cv.erode(img_bin, s_e2, iterations=3) #erosiona, a la resultante erosiona de nuevo y a la resultante la erosiona de nuevo

plt.imshow(np.hstack((img_bin, img_SE5, img_SE5_erodedx3)), "gray")
```

```python
img_bin_INV = cv.threshold(img, UMBRAL, 255, cv.THRESH_BINARY_INV)[1]

plt.imshow(cv.erode(img_bin_INV, s_e), "gray") #primer plano es negro, hace efecto contrario de erosion pq "quita info" de blanco
```

```python
img_bin_INV = cv.threshold(img, UMBRAL, 255, cv.THRESH_BINARY_INV)[1]

plt.imshow(cv.erode(img_bin_INV, s_e), "gray") #primer plano es negro, hace efecto contrario de erosion pq "quita info" de blanco
```

```python
img_bin_INV = cv.threshold(img, UMBRAL, 255, cv.THRESH_BINARY_INV)[1]

plt.imshow(cv.erode(img_bin_INV, s_e), "gray") #primer plano es negro, hace efecto contrario de erosion pq "quita info" de blanco
```

```python
img_dilatedx2 = cv.dilate(img_bin, s_e, iterations=2) #dilata, a la resultante dilata de nuevo

plt.imshow(np.hstack((img_bin, img_dilated, img_dilatedx2)), "gray")
```

```python
img = cv.imread('images\coins.png',0)

ret, imgf = cv.threshold(img,0,255,cv.THRESH_BINARY+cv.THRESH_OTSU)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(img, 'gray')
plt.title('Imagen Original')

plt.subplot(1, 2, 2)
plt.imshow(imgf, 'gray')
plt.title('Imagen Binarizada con Otsu')

plt.show()
```

```python
img = cv.imread('images\coins.png',0)

imgf = cv.adaptiveThreshold(img, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 11, 2)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(img, 'gray')
plt.title('Imagen Original')

plt.subplot(1, 2, 2)
plt.imshow(imgf, 'gray')
plt.title('Imagen con Umbralización Adaptativa')

plt.show()
```
