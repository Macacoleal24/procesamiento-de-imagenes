```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread("Mapache.jpeg",0)
plt.imshow(img, cmap="gray")
```

## Umbralizacin Global

```python
#Largo
height, weight = img.shape
matrix = np.zeros(img.shape)

for x in range(height):
    for y in range(weight):
        valor = img[x,y]
        if valor >= 128:
            matrix[x,y] = 255
        else: matrix[x,y] = 0

plt.imshow(matrix, cmap="gray")
```

```python
#Corta
imgbinaria = cv.threshold(img,128,1,cv.THRESH_BINARY)[1]
plt.imshow(imgbinaria, cmap="gray")
```

```python
#Invertida
plt.imshow(cv.threshold(img,128,1,cv.THRESH_BINARY_INV)[1], cmap="gray")
```

## Transformacin Neg

```python
#Formula s = L-1-r
#Largo

hght, wght = img.shape
s = np.zeros(img.shape)
for x in range(hght):
    for y in range(wght):
        r = img[x,y]
        s[x,y] = 255-1-r
plt.imshow(s, cmap="gray")
```

```python
#Corto
plt.imshow(np.bitwise_not(img,),"gray")
```

## Logaritmo

```python
m = np.max(img) #pixel ms grande

c = 255/(np.log(1+m))
alto, ancho = img.shape
s = np.zeros(img.shape)
for x in range(alto):
    for y in range (ancho):
        r = img[x,y]
        s[x,y]=c*np.log(1)+r
plt.imshow(s, cmap= "gray")
```

```python
C =255/np.log(1+m)
img_log = c*np.log(+img)
plt.imshow(img_log,"gray")
plt.show()
```

## Transformacin Gamma

```python
m = np.max(img)
pot =  25

alt, anch = img.shape
s =np.zeros(img.shape)
for x in range(alt):
    for y in range(anch):
        r = img[x,y]
        s[x,y] = (1)+r**pot

plt.imshow(s, cmap="gray")
```

```python
m=np.max(img)#sacandondo pixel mas grande
potencia=0.04

andlto_img, ancho_img = img.shape
s =np.zeros(img.shape)
for x in range(andlto_img):
    for y in range(ancho_img):
        r= img[x,y]  
        s[x,y]=(1)+r**potencia# CON LA GAMMA MAS PEQUA

plt.imshow(s,cmap='gray')
```
